```python
# use Python3
import pandas as pd
df_BTC = pd.read_pickle('BTC-USD-60.pkl')
df_BCH = pd.read_pickle('BCH-USD-60.pkl')
df_LTC = pd.read_pickle('LTC-USD-60.pkl')
df_ETH = pd.read_pickle('ETH-USD-60.pkl')
```

```python
import datetime
df_BTC['date'] = pd.to_datetime(df_BTC['time'],unit='s').dt.date
df_BTC = df_BTC[df_BTC['date'] >= datetime.date(2016,5,23)]

group_BTC = df_BTC.groupby('date')
BTC_Daily_Close_Price = group_BTC['close'].mean()

BTC_Daily_Close_Price=BTC_Daily_Close_Price.reset_index()
BTC_Daily_Close_Price=BTC_Daily_Close_Price.set_index('date')

BTC_Daily_Close_Price.head(1) # 2016-05-23	443.739544
BTC_Daily_Close_Price.shape #(681, 1)
```

```python
# 2016-05-21, 22 missing
import datetime
df_ETH['date'] = pd.to_datetime(df_ETH['time'],unit='s').dt.date
df_ETH = df_ETH[df_ETH['date'] >= datetime.date(2016,5,23)]

group_ETH = df_ETH.groupby('date')
ETH_Daily_Close_Price = group_ETH['close'].mean()

ETH_Daily_Close_Price=ETH_Daily_Close_Price.reset_index()
ETH_Daily_Close_Price=ETH_Daily_Close_Price.set_index('date')

ETH_Daily_Close_Price.head(1) # 2016-05-23	13.79
ETH_Daily_Close_Price.shape #(681, 1)
```

```python
# correlation
BTC_Daily_Close_Price.corrwith(ETH_Daily_Close_Price)
```

```python
# rolling correlation 
corrs5 = pd.rolling_corr(BTC_Daily_Close_Price["close"],ETH_Daily_Close_Price["close"], 5)
corrs10 = pd.rolling_corr(BTC_Daily_Close_Price["close"],ETH_Daily_Close_Price["close"], 10)
corrs90 = pd.rolling_corr(BTC_Daily_Close_Price["close"],ETH_Daily_Close_Price["close"], 90)
corrs180 = pd.rolling_corr(BTC_Daily_Close_Price["close"],ETH_Daily_Close_Price["close"], 180)
```

```python
import matplotlib.pyplot as plt
fig, axarr = plt.subplots(2, 2, figsize=(17,10))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                wspace=0.1, hspace=None)
fig.suptitle("BTC ETH Rolling Correlation", fontsize = 20)
axarr[0, 0].plot(corrs5)
axarr[0, 0].set_title("window_size = 5", fontsize = 18)
axarr[0, 1].plot(corrs10)
axarr[0, 1].set_title("window_size = 10", fontsize = 18)
axarr[1, 0].plot(corrs90)
axarr[1, 0].set_title("window_size = 90", fontsize = 18)
axarr[1, 1].plot(corrs180)
axarr[1, 1].set_title("window_size = 180", fontsize = 18)
```
