```python
import pandas as pd
df_BTC = pd.read_pickle('BTC-USD-60.pkl')

import datetime
df_BTC['date'] = pd.to_datetime(df_BTC['time'],unit='s').dt.date
df_BTC['datetime'] = pd.to_datetime(df_BTC['time'],unit='s').dt.time
df_BTC['hour'] = pd.to_datetime(df_BTC['time'],unit='s').dt.hour
Daily_BTC = df_BTC.groupby('date')

quiet_hours_BTC = Daily_BTC.apply(lambda x: x.loc[x['volume'] == x['volume'].min()])['hour'].values.tolist()
#quiet_time_BTC = Daily_BTC.apply(lambda x: x.loc[x['volume'] == x['volume'].min()])['datetime'].values.tolist()
```

```python
quietest_hour_BTC = max(set(quiet_hours_BTC), key=quiet_hours.count)
quietest_hour_BTC
```

```python
import pylab as pl

N, bins, patches = pl.hist(quiet_hours_BTC, bins = [x for x in range(24)])
pl.title("BTC Lowest-Voume Time")
pl.xlabel("Hour of day")
pl.ylabel("Count")

jet = pl.get_cmap('jet', len(patches))

for i in range(len(patches)):
    patches[i].set_facecolor(jet(i))
```
