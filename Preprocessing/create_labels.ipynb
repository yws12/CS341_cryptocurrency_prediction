```python
from scipy import stats
def find_bin(df,col_name,bin_num,is_pct_change):
    filled = df[col_name]
    if is_pct_change:
        filled = filled.pct_change()
        df[col_name + str("_pctChange")] = filled.copy()
        filled = filled.iloc[1:]
    print(filled.max())
    print(filled.min())
#     print(filled.isnull())
    bin_edges = stats.mstats.mquantiles(filled, [x/bin_num for x in range(bin_num + 1)])
    print(bin_edges)
    return bin_edges
```

```python
def create_label(df,col_name,need_pctChange):
    bin_edges = find_bin(df,col_name,3,need_pctChange)
    if (need_pctChange):
        labels = df[col_name+str("_pctChange")].copy()
    else:
        labels = df[col_name].copy()
    idx0 = (labels <= bin_edges[1])
    idx1 = (labels >= bin_edges[2])
    idx2 = ((labels < bin_edges[2]) & (labels>bin_edges[1]))
    labels[idx0] = 0
    print(len(labels[labels == 0]))
    labels[idx1] = 2
    print(len(labels[labels == 2]))
    labels[idx2] = 1
    print(len(labels[labels == 1]))
    df[col_name+str('_label')] = labels
```

```python
create_label(df,"USDT_BTC_pctChange",False)
# print()
create_label(df,"USDT_BTC_high", True)
create_label(df,"USDT_BTC_low", True)
create_label(df,"USDT_BTC_close", True)
create_label(df,"USDT_BTC_open",True)
create_label(df,"USDT_BTC_volatility", False)
df
```
