```python
__author__ = 'Yicheng Li'

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import statsmodels.api as sm
import random
import datetime

dir_path = 'Data/Poloniex/'

currency = 'USDT_BTC'
df = pd.read_csv(dir_path+currency+'.csv')
df = df.drop(columns='Unnamed: 0')
df = df.rename(columns={'date':'time'})
df['time'] = pd.to_datetime(df['time'],unit='s')
df.time = pd.DatetimeIndex(df.time) 
df = df.set_index('time')
df = df.sort_index()
df = df.rename(columns=\
        {'high':currency+'_high', 'low':currency+'_low', \
         'open':currency+'_open', 'close':currency+'_close', \
         'quoteVolume':currency+'_quoteVolume','volume':currency+'_volume',\
         'weightedAverage':currency+'_weightedAverage'})
DF = df

currency = 'BTC_ETH'
df = pd.read_csv(dir_path+currency+'.csv')
df = df.drop(columns='Unnamed: 0')
df = df.rename(columns={'date':'time'})
df['time'] = pd.to_datetime(df['time'],unit='s')
df.time = pd.DatetimeIndex(df.time) 
df = df.set_index('time')
df = df.sort_index()
df = df.rename(columns=\
        {'high':currency+'_high', 'low':currency+'_low', \
         'open':currency+'_open', 'close':currency+'_close', \
         'quoteVolume':currency+'_quoteVolume','volume':currency+'_volume',\
         'weightedAverage':currency+'_weightedAverage'})
DF = DF.join(df)

currency = 'BTC_LTC'
df = pd.read_csv(dir_path+currency+'.csv')
df = df.drop(columns='Unnamed: 0')
df = df.rename(columns={'date':'time'})
df['time'] = pd.to_datetime(df['time'],unit='s')
df.time = pd.DatetimeIndex(df.time) 
df = df.set_index('time')
df = df.sort_index()
df = df.rename(columns=\
        {'high':currency+'_high', 'low':currency+'_low', \
         'open':currency+'_open', 'close':currency+'_close', \
         'quoteVolume':currency+'_quoteVolume','volume':currency+'_volume',\
         'weightedAverage':currency+'_weightedAverage'})
DF = DF.join(df)

currency = 'BTC_XRP'
df = pd.read_csv(dir_path+currency+'.csv')
df = df.drop(columns='Unnamed: 0')
df = df.rename(columns={'date':'time'})
df['time'] = pd.to_datetime(df['time'],unit='s')
df.time = pd.DatetimeIndex(df.time) 
df = df.set_index('time')
df = df.sort_index()
df = df.rename(columns=\
        {'high':currency+'_high', 'low':currency+'_low', \
         'open':currency+'_open', 'close':currency+'_close', \
         'quoteVolume':currency+'_quoteVolume','volume':currency+'_volume',\
         'weightedAverage':currency+'_weightedAverage'})
DF = DF.join(df)
```
