```python
import pandas as pd
df = pd.read_pickle('df_daily_2.pickle')
df.head()
```

```python
df.columns
```

```python
import statsmodels.api as sm

tmp = df.dropna().iloc[:,[0, 1, 2, 4, 5, 6, 9, 11, 12, 16, 18, 19]]
tmp.head()
```

```python
import numpy as np

# normalize all features
normalized_df = tmp.transform(lambda x: (x - x.mean()) / x.std())

# shift all features 
shifted_df = normalized_df + 1 - normalized_df.min() 

# perform log transform on prices 
cols_of_interest = ['BTC_high', 'BTC_low', 'BTC_weighted_mean', 'ETH_weighted_mean', 'LTC_weighted_mean']
shifted_df[cols_of_interest] = np.log(shifted_df[cols_of_interest])

transformed_df = shifted_df
transformed_df.head()
```

```python
from datetime import date
train_start_date = date(2016,8,18)
test_end_date = date(2018,4,3)

total_days = (test_end_date - train_start_date).days + 1
print(total_days)
```

```python
num_days_train, num_days_validate = int(total_days * 0.7), int(total_days * 0.2)
num_days_test = total_days - num_days_train - num_days_validate
print(num_days_train, num_days_validate, num_days_test)
```

```python
trainX = transformed_df.iloc[:415].as_matrix()
trainX.shape
```

```python
trainY = transformed_df.iloc[1:416, 1].as_matrix()
trainY.shape
```

```python
import numpy as np

# define evaluation metric
def SMAPE(y, y_pred):
    if len(y) != len(y_pred):
        raise ValueError('Length of prediction array is not equal to length of y array.')
    return np.mean(np.abs(y-y_pred)*2/(np.abs(y)+np.abs(y_pred)))
```

```python
import statsmodels.api as sm

aic = []
bic = []
Train_SMAPE = []
Dev_SMAPE = []

# Full model
model = sm.OLS(trainY, trainX).fit()
aic.append(model.aic)
bic.append(model.bic)
# Print out the statistics
print(model.summary())
```

```python
trainY_pred = model.predict(trainX) # make the predictions by the model
train_SMAPE = SMAPE(trainY, trainY_pred)
print('train:', train_SMAPE)

devX = transformed_df.iloc[415:415+118].as_matrix()
devY = tmp.iloc[416:416+118, 1].as_matrix()

devY_pred = model.predict(devX) # make the predictions by the model
dev_SMAPE = SMAPE(devY, devY_pred)
print('dev:', dev_SMAPE)

Train_SMAPE.append(train_SMAPE)
Dev_SMAPE.append(dev_SMAPE)
```
